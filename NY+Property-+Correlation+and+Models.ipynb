## Linear Regression

```python
import statsmodels.formula.api as smf
```

```python
df["ZIP_CODE"]=df["ZIP CODE"]
df["YEAR_BUILT"]=df["YEAR BUILT"]
df["RESIDENTIAL_UNITS"]=df["RESIDENTIAL UNITS"]
df["COMMERCIAL_UNITS"]=df["COMMERCIAL UNITS"]
```

```python
result = smf.ols('ZIP_CODE ~ YEAR_BUILT + RESIDENTIAL_UNITS + COMMERCIAL_UNITS', data=df).fit()
```

```python
result.summary()
```

With age, exercise, and hours being 0, your starting grade is likely to be around 57.87

```python
#remove age from regression, not very correlated
result = smf.ols('grade ~ exercise + hours', data=df).fit()
result.summary()
```

```python
#set coefficient to 0
result = smf.ols(formula='grade ~ age + exercise + hours - 1', data=df).fit()
result.summary()
```
